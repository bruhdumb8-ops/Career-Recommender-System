```python
# AI Career Recommender System

This project is a simple Python-based Career Recommender that suggests possible career paths
based on a user's preferences and interests.

It demonstrates:
- Conditional logic in Python
- Mapping user input to career recommendations
- Applying basic problem-solving skills in code

## Instructions

- Answer all questions with "yes" or "no".
- The program will suggest a career based on your answers.
- It will also give a suggested next step for learning or preparation.

# Career Recommender System

def recommend_career():
    print("Let's find the best career for you! Answer the following questions with yes or no.\n")

    # User inputs
    likes_data = input("Do you like working with data? ").lower()
    likes_people = input("Do you like working with people? ").lower()
    likes_ideas = input("Do you enjoy thinking about ideas/concepts? ").lower()
    teamwork = input("Do you like working in a team? ").lower()
    likes_challenges = input("Do you enjoy solving challenging problems? ").lower()

    # Convert yes/no to numbers
    likes_data = 1 if likes_data == "yes" else 0
    likes_people = 1 if likes_people == "yes" else 0
    likes_ideas = 1 if likes_ideas == "yes" else 0
    teamwork = 1 if teamwork == "yes" else 0
    likes_challenges = 1 if likes_challenges == "yes" else 0

    # Decide career based on simple rules
    if likes_data and not likes_people and likes_challenges:
        career = "Data Scientist"
    elif likes_people and not likes_data:
        career = "Psychologist"
    elif likes_ideas and not likes_data and not likes_people:
        career = "Philosopher"
    elif likes_data and likes_people and likes_challenges:
        career = "Software Engineer"
    elif likes_data and likes_ideas and likes_challenges:
        career = "AI Researcher"
    elif likes_people and likes_ideas:
        career = "Counselor"
    else:
        career = "General Analyst"

    # Suggest next steps
    next_steps = {
        "Data Scientist": "Learn Python, SQL, and basic Machine Learning.",
        "Psychologist": "Consider studying Psychology or Counseling.",
        "Philosopher": "Read about Philosophy and Critical Thinking.",
        "Software Engineer": "Learn Python, Java, and Web Development.",
        "AI Researcher": "Study AI concepts and Neural Networks.",
        "Counselor": "Take courses in Communication and Psychology.",
        "General Analyst": "Practice Excel, data analysis, and logical thinking."
    }

    # Output recommendation
    print(f"\nRecommended Career: {career}")
    print(f"Suggested Next Step: {next_steps[career]}")

# Run the program
recommend_career()

## Example Run

If the user answers:
- Do you like working with data? yes
- Do you like working with people? no
- Do you enjoy thinking about ideas/concepts? yes
- Do you like working in a team? yes
- Do you enjoy solving challenging problems? yes

The output will be:

Recommended Career: Data Scientist
Suggested Next Step: Learn Python, SQL, and basic Machine Learning.
```
